# Atividade 2
Um exemplo de um sinal bioelétrico é o eletrocardiograma (ECG), que mede a atividade elétrica do coração ao longo do tempo. Ele é utilizadono contexto da biomedicina  para diagnosticar problemas cardíacos e monitorar a saúde cardiovascular do paciente. Ele tem como sua origem o coração que funciona através da geração e propagação de impulsos elétricos. O ECG detecta a variação da voltagem gerada pelos impulsos elétricos que realizam as contrações cardíacas.

Esse sinal apresenta como características a onda P que representa a despolarização atrial, ou seja, a contração dos átrios. O Complexo QRS que indica a despolarização ventricular, que é a contração dos ventrículos.
Segmento ST: Reflete o intervalo entre a despolarização e repolarização ventricular. A onda T que representa a repolarização ventricular. A onda U e T que nem sempre são registradas no ECG. O intervalo PR, o qual é o intervalo onde a onda P inicia e o início do complexo QRS. Período PP que é o intervalo entre duas ondas P. Por fim temos o Período RR que é o intervalo entre duas ondas R.

Suas aplicações vão da avaliação das condições cardíacas como diagnóstico de arritmias identificando irregularidades no ritmo cardíaco, como taquicardia e bradicardia. Detecção de infarto do miocárdio onde alterações específicas no ECG podem vir a indicar danos ao músculo.

## Questão 1: Uma rotina em python que gere uma senoide de Amplitude A, frequência omega e fase theta

```python
import numpy as np
# Questão 1: Uma rotina em python que gere uma senoide de Amplitude A, frequência omega e fase theta
def senoide (a, omega, theta, t):
    x = A * np.sin(omega * t + theta)
    senoide_dict = {}
    senoide_dict = {
        'nome': 'senoide',
        'a': A,
        'omega': omega,
        'theta': theta,
        't': t,
        'x': x
    }
    return senoide_dict
''' Gera os pontos de uma senoide em nos valores assumidos em t, com os parâmetros dados:
    * a: amplitude (volt);
    * omega: freq angular (rad/s);
    * theta: fase

  Retorna:
    * Dicionário com os seguintes campos:
      * 'nome': 'senoide'
      * 'a': valor do parâmetro de entrada 'a';
      * 'omega': valor do parâmetro de entrada 'omega';
      * 'theta': valor do parâmetro de entrada 'fase';
      * 't': valores do parâmetro de entrada 't';
      * 'x': valores da senoide gerada;
'''
```
